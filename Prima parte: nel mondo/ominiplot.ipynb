# Plot con gli omini
Per rappresentare le differenze tra umoni e donne laureate (in alcuni paesi) abbiamo deciso di generare un plot "pittogramma" composto di una quantità di omini rappresentati donne e uomini nelle quantità che rispecchiano le loro percentuali.

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
```

```python
from collections import defaultdict
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

with open('laureati_nel_mondo_field.csv', 'r') as data_file:
  data_reader = csv.reader(data_file, delimiter =',', quotechar ='"')
  data = list(data_reader)[1:]

data = pd.read_csv('laureati_nel_mondo_field.csv', sep=',', index_col=0)
data = data.reset_index()

sdata = data[['Country', 'SEX', 'Field', 'Education level', 'Year', 'Value']]

scdata = sdata.dropna(subset=['Value'])
scdata
```

```python
NMS = sdata[sdata['Field'] == "Natural sciences, mathematics and statistics"]
ICT = sdata[sdata['Field'] == "Information and Communication Technologies (ICTs)"]
EMC = sdata[sdata['Field'] == "Engineering, manufacturing and construction"]

STEMdata = pd.concat([NMS, ICT, EMC])
STEMdata = STEMdata.groupby(['Country', 'SEX', 'Year'])['Value'].sum()
STEMdata = STEMdata.reset_index()

STEMdata21 = (STEMdata[STEMdata['Year'] == 2021])

STEMdataUSA = STEMdata21.query('Country == "United States"')

Tot = STEMdataUSA.groupby(['Country', 'Year'])['Value'].sum()
percentF = STEMdataUSA.query('SEX == "F"')['Value'] 
percentF = (int(percentF) / int(Tot))*100
percentM = STEMdataUSA.query('SEX == "M"')['Value'] 
percentM = (int(percentM) / int(Tot))*100

percentF = round(percentF)
percentM = 100 - percentF

# Waffle chart
plt.figure(
    FigureClass = Waffle,
    rows = 10,
    title = {"label": "USA", "loc": "center", "size": 15},
    values = [percentM,percentF],
    icons = ['person','person-dress'],
    colors =["#8aadd0","#ff7088"],
    facecolor = '#DAD1D5',
    vertical = 'True')
#5C51F6#B6CBE2

plt.show()
```
