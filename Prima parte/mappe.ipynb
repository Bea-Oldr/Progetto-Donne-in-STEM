# Mappe cloroplete
Abbiamo deciso di presentare l'informazione della quantità di donne laurate al mondo in forma di mappa cloropleta, le mappe qui elaborate rappresentano il quantitativo totale di laureate in ogni paese e il quantitativo di donne laureate specificamente i materie STEM di ogni paese. Sono state elaborate un totale di 4 mappe, due per ogni database a disposizione, e sono state scelte le mappe relative al database più completo.

```python
from collections import defaultdict
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
```

## Primo database

```python
with open('Share of graduates by field, female (%).csv', 'r') as data_file:
  data_reader = csv.reader(data_file, delimiter =',', quotechar ='"')
  data = list(data_reader)[1:]

data = pd.read_csv('Share of graduates by field, female (%).csv', sep=',', index_col=0)

clear_data = data[['Country Name', 'Year', 'Value']].reset_index()

C = clear_data.query('Year > 2007 & Year < 2017')
Cgrouped = C.groupby(['Country Name'])['Value'].sum()
Cgrouped = Cgrouped.reset_index()

fig = px.choropleth(Cgrouped, locations='Country Name', locationmode='country names', color='Value', projection='natural earth', title='Laureate per paese 2007 - 2017', 
                    range_color = (0, 5000), color_continuous_scale='ice_r', width=700)

fig.show()
```

```python
C = clear_data[clear_data["Indicator Name"] == "Female share of graduates from Science, Technology, Engineering and Mathematics (STEM) programmes, tertiary (%)"]
S = clear_data[clear_data["Indicator Name"] == "Female share of graduates in Natural Sciences, Mathematics and Statistics programmes, tertiary (%)"]   
T = clear_data[clear_data["Indicator Name"] == "Female share of graduates in Engineering, Manufacturing and Construction programmes, tertiary (%)"]             

CSTEM = pd.concat([C,S,T])                 

CCSTEM = CSTEM.query('Year > 2007 & Year < 2017')
CCSTEMgrouped = CCSTEM.groupby(['Country Name'])['Value'].sum()
CCSTEMgrouped = CCSTEMgrouped.reset_index()

fig = px.choropleth(CCSTEMgrouped, locations='Country Name', locationmode='country names', color='Value', projection='natural earth', title='Laureate stem per paese 2007 - 2017', 
                    range_color = (0, 5000), color_continuous_scale='ice_r', width=700)

fig.show()
```

## Secondo database

```python
with open('laureati_nel_mondo_field.csv', 'r') as data_file:
  data_reader = csv.reader(data_file, delimiter =',', quotechar ='"')
  data = list(data_reader)[1:]

data = pd.read_csv('laureati_nel_mondo_field.csv', sep=',', index_col=0)
data = data.reset_index()

sdata = data[['Country', 'SEX', 'Field', 'Education level', 'Year', 'Value']]

scdata = sdata.dropna(subset=['Value'])
scdata
```

```python
Q = scdata.query('SEX == "F"', inplace=True)
Q = scdata.query('Year == 2021')
newQ = Q

fig = px.choropleth(newQ, locations='Country', locationmode='country names', color='Value', projection='natural earth', title='Laureate per paese 2013 - 2021', 
                    range_color = (0, 50000), color_continuous_scale=('PowderBlue', 'Blue', 'SlateBlue'))

fig.show()
```

```python
NMS = sdata[sdata['Field'] == "Natural sciences, mathematics and statistics"]
ICT = sdata[sdata['Field'] == "Information and Communication Technologies (ICTs)"]
EMC = sdata[sdata['Field'] == "Engineering, manufacturing and construction"]

STEMdata = pd.concat([NMS, ICT, EMC])
STEMdata = STEMdata.groupby(['Country', 'SEX', 'Year'])['Value'].sum()
STEMdata = STEMdata.reset_index()

fig = px.choropleth(STEMdata, locations='Country', locationmode='country names', color='Value', projection='natural earth', title='Laureate per paese', 
                    range_color = (0, 50000), color_continuous_scale=('PowderBlue', 'Blue', 'SlateBlue'))

fig.show()
```
